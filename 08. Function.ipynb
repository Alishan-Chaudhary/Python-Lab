```python
def adder(num_1, num_2):
  """
  Adds the number

  Args:
    num_1 (float): First number
    num_2 (float): Second number

  Returns:
    float: Sum of the two numbers
  """
  total = num_1 + num_2
  diff = num_1 - num_2
  return total, diff
```

```python
tot, di = adder(4, 3)
print(tot)
print(di)
```

```python
def incrementor(var_1, increment=1):
  new_value = var_1 + increment
  return new_value
```

```python
print(incrementor(5, 2))
print(incrementor(5))
```

```python
def kwargs_unformat(**kwargs):
  for key, value in kwargs.items():
    print(f"{key} = {value}")
```

```python
kwargs_unformat(num_1=4)
print()
kwargs_unformat(num_1=4, num_2=3)
```

```python
def factorial(n):
  if n == 1:
    return 1
  else:
    return n * factorial(n-1)
```

```python
print(factorial(3))
```

```python
sqr_lambda = lambda x, y: x*y
print(sqr_lambda(3,5))
```

```python
# Filter
user_age = [13, 15, 16, 18]
filtered_age = list(filter(lambda x: x>=16, user_age))
print(filtered_age)
```

```python
# Map
movies_list = [' Thor ', 'Avengers ']
cleaned_list = list(map(lambda x:x.strip().upper(), movies_list))
print(cleaned_list)
```
